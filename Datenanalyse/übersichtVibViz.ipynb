```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import os
import json

json_path = "..\\viblib\\test\\2\\json"

#Lädt json Datei und gibt Inhalt als string zurück
def readJson(file_name):
  file_path = os.path.join(json_path, file_name + ".json")
  if not os.path.isfile(file_path):
    raise Exception("File", file_path, "does not exists.")
  jsonObj = None
  with open(file_path,'r') as file:
    jsonObj = json.load(file)
  return jsonObj

def getJsonNames():
    # Liste, um die .json-Dateinamen zu speichern
    jsonFileNames = []

    # Durchlaufe alle Dateien im angegebenen Ordner
    for file in os.listdir(json_path):
        # Prüfe, ob die Datei die Endung .json hat
        if file.endswith(".json"):
            name=os.path.splitext(file)[0]
            jsonFileNames.append(name)

    return jsonFileNames

names = getJsonNames()

```

## Tabelle Blockanzahl, Dauer

```python
import numpy
def countBlocksWithType(jsonObj,type):
    count = 0
    for block in jsonObj["sequence"]:
        if block["type"] == type:
            count += 1
    return count
def countBlocks(jsonObj):
    return len(jsonObj["sequence"])
def duration(jsonObj):
    dur = 0
    for block in jsonObj["sequence"]:
        dur += block["data"]["dur"]
    return dur

data = {
    "ID" : [],
    "Total block count" : [],
    "Breaks" : [],
    "Vibration blocks" : [],
    "Duration in ms" : []
}
for name in names:
    jsonObj = readJson(name)
    data["ID"].append(name)
    data["Total block count"].append(countBlocks(jsonObj))
    data["Vibration blocks"].append(countBlocksWithType(jsonObj,"v"))
    data["Breaks"].append(countBlocksWithType(jsonObj,"p"))
    data["Duration in ms"].append(duration(jsonObj))

data["ID"].append("mean")
data["Total block count"].append(numpy.mean(data["Total block count"]))
data["Vibration blocks"].append(numpy.mean(data["Vibration blocks"]))
data["Breaks"].append(numpy.mean(data["Breaks"]))
data["Duration in ms"].append(numpy.mean(data["Duration in ms"]))


df = pd.DataFrame(data)
pd.set_option("display.max_rows", None)
df
```
